Loading the training dataframe

```python
# load the training csv
# convert id column datatype to string 
df_train_full = pd.read_csv('data/train.csv', dtype={'Id': str})
# making another column (named filename)
## this contains the images file path using id's in the train csv 
df_train_full['filename'] = 'data/images/' + df_train_full['Id'] + '.jpg'
df_train_full.head()
```

Splitting the loaded train dataset to:

- Training dataset: 80%
- Validation dataset: 20%

```python
val_cutoff = int(len(df_train_full) * 0.8)
# train dataset
df_train = df_train_full[:val_cutoff]
# validation dataset 
df_val = df_train_full[val_cutoff:]
```

```python
# checking the shape of the splited data

df_train.shape, df_val.shape
```

Creating image generator for which the model will be trained on 

```python
## importing the module to read the images
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# read and generate the train and validation data 
## Note: preprocessing_function parameter will be set to preprocess_input module
## used for Xception model

############## Training Data ####################
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_gen = train_datagen.flow_from_dataframe(
    df_train,
    x_col='filename',
    y_col='label',
    target_size=(150, 150),
    batch_size=32,
)

############ Validation Data ####################
val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

val_gen = val_datagen.flow_from_dataframe(
    df_val,
    x_col='filename',
    y_col='label',
    target_size=(150, 150),
    batch_size=32,
)
```

Let's confirm the classes of the image generated for training set 

```python
train_gen.class_indices
```

I want train initialize the base model with it's accompany parameters.

**`Things to Note:`**
- Xception is the keras application that we intend adopting 
- Since the Xception was trained on imagenet dataset, weights paratemer will be set to 'imagenet'
- The project focuses on transfer learning, so, in the Xception model, its Dense layers will be removed by setting `include_top` parameter as `False`
- Input_shape of the images will be set to `(150, 150, 3)`:- help training the model faster 
- For the model not be trained all over again, the base model will be set to `False`
- The output of the convolutional layers of the base model will be transform to Vector Representation using `Pooling`
- Then the output of the Dense Layers (which we intend training) will be set to `6` (since the project classes is 6) using Transform Vectors as input  

```python
# base model
base_model = Xception(
    weights='imagenet',
    include_top=False,
    input_shape=(150, 150, 3)
)
# not training the model all over
base_model.trainable = False

# input dependency
inputs = keras.Input(shape=(150, 150, 3))

base = base_model(inputs, training=False)
# converting the CL's to Vector representation
vectors = keras.layers.GlobalAveragePooling2D()(base)
# using the vector representation to train a dense layer 
outputs = keras.layers.Dense(6)(vectors)

model = keras.Model(inputs, outputs)
```

**Note:** Activation function (e.g SOFTMAX) will not be added, since we want the output not in probability. 

```python
learning_rate = 0.01

######### Model Optimizer ############
# Using Adam optimizer
# with the set learning rate 
optimizer = keras.optimizers.Adam(learning_rate=learning_rate)

######## Model Loss #########
# Using CategoricalCrossentropy
# Setting logits to True, since categorial result not probability
loss = keras.losses.CategoricalCrossentropy(from_logits=True)

######## Model Compiler #########
# Compiling the model with optimizer, loss, and metrics evaluation 
model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
```

```python
# fitting the model 
# training the model on 10 epochs

history = model.fit(
    train_gen,
    epochs=10,
    validation_data=val_gen
)
```

## Experimenting the model learning rate with different values as against the model performance on each learning rate value (best learning rate will be chosen)


```python
# defining a function 

def make_model(learning_rate=0.01):
  base_model = Xception(
      weights='imagenet', 
      include_top=False, 
      input_shape= (150, 150, 3)
  )

  base_model.trainable = False

  ################################################

  inputs = keras.Input(shape=(150, 150, 3))
  base = base_model(inputs, training=False)
  vectors = keras.layers.GlobalAveragePooling2D()(base)
  outputs = keras.layers.Dense(6)(vectors)
  model = keras.Model(inputs, outputs)

  ################################################

  optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
  loss = keras.losses.CategoricalCrossentropy(from_logits=True)

  model.compile(optimizer=optimizer, 
                loss=loss, 
                metrics=['accuracy'])

  
  return model
```

```python
# iterating over different values of learning rate 

scores = {}

for lr in [0.0001, 0.001, 0.01, 0.1]:
  print(lr)

  model = make_model(learning_rate=lr)
  history = model.fit(train_gen, epochs=10, validation_data=val_gen)
  scores[lr] = history.history

  print()
  print()
```

##  Experiment by adding inner dense layers to the networks 

The idea behind this is to add another layer (inner dense layer) between the input vector representation and the output dense layer. What this does is to do some intermediate processing of the input vector represention, and the output of this layer will be transmitted to output dense layer. 

We want the model to be more powerful and learn some internal representation from its supplied input (VR)

**Note:** this can be disregarded if it adds nothing to the model performance.

To-do:
- Using different sizes of inner dense layers (best combination will be chosen)
- Setting the learning rate to `0.001`
- Add activation function (in this case `RELU`(Rectified Linear Unit)) to the inner layer. This will process the output of the dense vector
- Experiment by regularization and dropout 
- Visualize the performance 
- Checkpoint the best combination

```python
# defining a function 

def make_model(learning_rate=0.001, size_inner=100):
  base_model = Xception(weights='imagenet', 
                        include_top=False, 
                        input_shape= (150, 150, 3))

  base_model.trainable = False

  ############## Vector Representation ###########################

  inputs = keras.Input(shape=(150, 150, 3))
  base = base_model(inputs, training=False)
  vectors = keras.layers.GlobalAveragePooling2D()(base)

  ############## Inner Dense Layer #################

  inner = keras.layers.Dense(size_inner, activation='relu')(vectors)

  ############## Output ##################

  outputs = keras.layers.Dense(6)(inner)
  model = keras.Model(inputs, outputs)

  ################################################

  optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
  loss = keras.losses.CategoricalCrossentropy(from_logits=True)

  model.compile(optimizer=optimizer, 
                loss=loss, 
                metrics=['accuracy'])

  
  return model
```

```python
# tuning the size of the layers 

learning_rate = 0.001

scores = {}

for size in [10, 100, 1000]:
  print(size)

  model = make_model(learning_rate=learning_rate, size_inner=size)
  history = model.fit(train_gen, epochs=10, validation_data=val_gen)
  scores[size] = history.history

  print()
  print()
```

# Data Augmentation

The idea behind Data Augmentation is creating more data from the existing dataset. This will determine if creating more data from the existing dataset can help improve the performance of the model. 

You can remember that we deployed dropout hyperparameter tuning in regularising the model to avoid the model to overfit. Another technique that can be used inplace of dropout is Data Augmentation. This avoid cases where the model would see similar images while training, and also defeat cases where the model would memorize. 

To do this, I'll be following the below image transformation guidelines:
- Flip ==> horizontal, vertical flipping, or both. 
- Rotation 
- Shifting ==> up, down, right, and left 
- Shear 
- Zoom (in/out)
- Brightness/contrast 

The augmentation techniques that would be used would based on image usecase for this project 

**Note:** I'll investigate each augmentation techniques then combine them altogether

**TENSORFLOW LITE** 

**Why the choice the TensorFlow Lite**

1. For the accompany limits of AWS Lambda, the whole TensorFlow package can't be used because of its size. This prompts the use of TensorFlow Lite which is smaller in size.

2. Using TensorFlow leads to Large Docker Image, which will results in slower running time, and payment of resources used within AWS Lambda Service 

Liter version of TensorFlow (TensorFlow Lite) focuses more on **inference (model.predict(X))**. Unlike the larger version TensorFlow, TF-Lite can't be used for training the model, but focus more on making prediction (inference). This makes the prediction runtime faster. 

To use this, I'll be transforming the model I built earlier to smaller size. 

Follow through as I transform the model 

**`1. Making a normal sample prediction before conversion to tensorflow lite`**

```python
# importing the libraries 

import numpy as np

import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.applications.xception import preprocess_input
```

```python
# loading the model 

model = keras.models.load_model('xception_v2_299_15_0.967.h5')
```

```python
# loading the sample image and process it 
img = load_img('images/0096.jpg', target_size=(299, 299))

x = np.array(img)
X = np.array([x])

X = preprocess_input(X)
```

```python
# making prediction
pred = model.predict(X)
```

```python
# prediction classes 
classes = ['cup', 
           'forks', 
           'glass',
           'knife', 
           'plate', 
           'spoon']
```

```python
# combining label classes and model prediction

dict(zip(classes, pred[0]))
```

This is using the whole TensorFlow package to make the prediction. Now, we want to use TensorFlow lite to make the prediction by using its INFERENCE. 

**`2. Converting the model from tensorflow to tensorflow lite`** 

This converts the keras application model to TF-Lite format.

```python
# converting the model to lite 
converter = tf.lite.TFLiteConverter.from_keras_model(model)

tflite_model = converter.convert()

# saving the model in tensorflow lite format
with open('kitchenware-model.tflite', 'wb') as f_out:
    f_out.write(tflite_model)
```

**`Note: Everything here will be done outside the purview of Full Version TensorFlow Package`**

```python
# importing the frameworks 
import tflite_runtime.interpreter as tflite
from keras_image_helper import create_preprocessor
```

```python
# interpreting the model
interpreter = tflite.Interpreter(model_path='kitchenware-model.tflite')
# initialize the model
interpreter.allocate_tensors()

# get the input index of the interpreter 
input_index = interpreter.get_input_details()[0]['index']
# get the output index of the interpreter
output_index = interpreter.get_output_details()[0]['index']
```

```python
# get model architecture 
preprocessor = create_preprocessor('xception', target_size=(299, 299))
```

```python
# preprocessing the image 
url = 'https://bit.ly/3IYiPAX'
X = preprocessor.from_url(url)
```

```python
# supplying the model with sample image for prediction
interpreter.set_tensor(input_index, X)
interpreter.invoke()
preds = interpreter.get_tensor(output_index)
```

```python
# prediction classes 
classes = ['cup', 
           'forks', 
           'glass',
           'knife', 
           'plate', 
           'spoon']

# combining label classes and model prediction
dict(zip(classes, preds[0]))
```

## How the Service Works for the Solution Provider and Receiver

For the service to benefit the user, it's imperative for the solution provider to create a software solution around the **API**. 

**API workflow**
- A user uploads a kitchenware image (containing the classes for which the model was trained with) to the solution provider platform
- A request will be sent through the API gateway using the uploaded image 
- API gateway transmits this request to the lambda function 
- The lambda function receives this request and make prediction
- Lambda function sends its response through API gateway 
- API gateway receives and transmits the lambda function response
- Then the user receives the response 

All these happen within seconds of request by the user. 

**Note:** this is beneficial to the user after a software solution is created around the API
